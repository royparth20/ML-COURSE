```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)
```

```python
plt.subplot(1,2,1)
plt.plot(x1,y1,'.-')
plt.title('1')

plt.subplot(1,2,2)
plt.plot(x2,y2,'o-')
plt.title('2')
plt.show()
```
