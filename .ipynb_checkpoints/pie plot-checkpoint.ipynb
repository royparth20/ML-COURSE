```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
activites = ['eat','sleep','work','play']
slices = [3,7,8,6]
colors = ['r','g','m','b']
```

```python
plt.pie(slices, labels=activites, colors=colors,startangle=90,shadow=True,explode=(0.2,0,0.1,0),autopct='%1.2f%%')

plt.legend()
plt.show()
```
