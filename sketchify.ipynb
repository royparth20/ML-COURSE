```python
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```python
import numpy as np
def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
img ="D:\Python\ML HINDI COURSE\images\\nilkanth.jpg"
```

```python
import imageio
s = imageio.imread(img)
g=grayscale(s)

import matplotlib.pyplot as plt

plt.subplot(221)
plt.imshow(g)

i = 255-g
plt.subplot(222)
plt.imshow(i)

import scipy.ndimage
b = scipy.ndimage.filters.gaussian_filter(i,sigma=10)
plt.subplot(223)
plt.imshow(b)

r= dodge(b,g)
plt.subplot(224)
plt.imshow(r)
```
