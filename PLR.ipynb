```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data=pd.read_csv("Datasets/Position_Salaries.csv")
data.head(10)
```

```python
X=data.iloc[:,1:2].values
Y=data.iloc[:,2].values
print(X,Y)
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
polynomial_reg=PolynomialFeatures(degree=4)
X_poly=polynomial_reg.fit_transform(X)
X_poly
```

```python
polynomial_reg.fit(X_poly,Y)
linear_reg=LinearRegression()
linear_reg.fit(X_poly,Y)
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,model.predict(X),color="green")
plt.show()
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,linear_reg.predict(polynomial_reg.fit_transform(X)),color="green")
plt.show()
```

```python
linear_reg.predict(polynomial_reg.fit_transform([[6.5]]))
```
