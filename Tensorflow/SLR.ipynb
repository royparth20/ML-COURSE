```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution()
import numpy as np
```

```python
X = np.linspace(0,10,1000)+np.random.uniform(-1.5,1.5,1000)
Y = np.linspace(0,10,1000)+np.random.uniform(-2.5,2.5,1000)
# print(X,Y)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,Y,"*")
```

```python
# y=mx+c
```

```python
np.random.rand(2)
```

```python
m = tf.Variable(0.58)
b = tf.Variable(0.98)
```

```python
#Cost Function

error = 0
for x,y in zip(X,Y):
    y_pred= m*x + b
    error = error + (y-y_pred)**2
```

```python
# print(error)
# session = tf.compat.v1.InteractiveSession()
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.001)
train = optimizer.minimize(error)
```

```python
init = tf.compat.v1.global_variables_initializer()
```

```python
with tf.compat.v1.Session() as session:
    session.run(init)
    no_time_train = 1
    for i in range(no_time_train):
        session.run(train)
    slope, intercept = session.run([m,b])
```

```python
x_test=np.linspace(0,1,50)
y_pred_plot = slope * x_test + intercept 
print(x_test,y_pred_plot)
```
